```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 10)
```

```python
y = x**2
```

```python
plt.plot(x, y, "r--")
plt.xlim(0, 4)
plt.ylim(0, 10)
plt.title("TITLE")
plt.xlabel("X label")
plt.ylabel("Y label")
```
